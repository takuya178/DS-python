```python
df2 = pd.read_csv('dataset/tmdb_5000_movies.csv')
df = df2[~df2['release_date'].isna()]
df['revenue_year'] = df['release_date'].apply(lambda x: int(x[:4]))

groupby_revenue = df.groupby('revenue_year')
median_revenue = groupby_revenue.median()['revenue']
max_revenue = groupby_revenue['revenue'].max()
mean_revenue = groupby_revenue.mean()['revenue']

plt.plot(median_revenue, label= 'median')
plt.plot(max_revenue, label='max')
plt.plot(mean_revenue, label='mean')
plt.legend()
```

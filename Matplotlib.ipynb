```python
df2 = pd.read_csv('dataset/tmdb_5000_movies.csv')
df = df2[~df2['release_date'].isna()]
df['revenue_year'] = df['release_date'].apply(lambda x: int(x[:4]))

groupby_revenue = df.groupby('revenue_year')
median_revenue = groupby_revenue.median()['revenue']
max_revenue = groupby_revenue['revenue'].max()
mean_revenue = groupby_revenue.mean()['revenue']

plt.plot(median_revenue, label= 'median')
plt.plot(max_revenue, label='max')
plt.plot(mean_revenue, label='mean')
plt.legend()
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df.head()
```

```python
plt.scatter(df['revenue'], df['budget'], alpha=0.2)
plt.xlabel('revenue')
plt.ylabel('budget')
```

```python
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
lang_group_df = df.groupby('original_language').mean()[['budget', 'revenue', 'vote_count']]
c = np.arange(len(lang_group_df))
plt.scatter('revenue', 'budget', s='vote_count', data=lang_group_df, c=c)
plt.xlabel('revenue')
plt.ylabel('budget')

for idx, row in lang_group_df.iterrows():
  plt.annotate(idx, (row['revenue'], row['budget']))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
sns.distplot(df['vote_count'])
```

```python
sns.jointplot(x="vote_count", y="vote_average", data=df, alpha=0.2)
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df = df[df['vote_count'] > 100]
df

sns.jointplot(x='vote_count', y='vote_average', data=df, kind='reg', joint_kws={'scatter_kws': {'alpha': 0.3}})
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
mean_vote_avrage = df['vote_average'].median()
df['evaluation'] = df['vote_average'].apply(lambda x: "high" if x > mean_vote_avrage else "low")

sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average', 'evaluation']], hue='evaluation')
```
